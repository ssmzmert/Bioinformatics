```python
import numpy as np
```

```python
input_file = 'global_aln_test.txt' 

def print_matrix(input_file, A):
    """Print the matrix with the (0,0) entry in the top left
    corner. Will label the rows by the sequence and add in the
    0-row if appropriate."""
    with open(input_file,'r') as f:
        sequences = f.readlines()
        x = sequences[0].strip()
        y = sequences[1].strip()
    f.close()
    # decide whether there is a 0th row/column
    if len(x) == len(A):
        print ("%5s" % (" "),end = '')
    else:
        print ("%5s %5s" % (" ","*"),end = '')
        y = "*" + y

    # print the top row
    for c in x:
        print ("%5s" % (c),end = '')
    print()

    for j in range(len(A[0])):
        print ("%5s" % (y[j]),end = '')
        for i in range(len(A)):
            print ("%5.0f" % (A[i][j]),end = '')
        print()

def global_align(input_file, gap_opening_penalty, gap_extend_penalty,match_score,mismatch_score):
    f = open(input_file,'r')
    sequences = f.readlines()
    x = sequences[0].strip()
    y = sequences[1].strip()
    f.close()

    # TO-DO 2: Create the matrix
    matrix = np.zeros((len(x)+1,len(y)+1))


# TO-DO 3: Fill up the score matrix
    tmp = gap_opening_penalty
    for i in range(len(x)):
     matrix[i+1][0] = tmp
     tmp += gap_opening_penalty
    tmp = gap_opening_penalty
    for i in range(len(y)):
       matrix[0][i+1] = tmp
       tmp += gap_opening_penalty


    for i in range(len(x)):
      for k in range(len(y)):
        if x[i] == y[k]:
         diagonal = matrix[i][k] + match_score
        else:
         diagonal = matrix[i][k] + mismatch_score
        vertical = matrix[i+1][k] + gap_opening_penalty
        horizontal = matrix[i][k+1] + gap_opening_penalty
        num = max(diagonal,vertical,horizontal)
        matrix[i+1][k+1] = num


# TO-DO 4: Find the maximum score and its indexes 
    max_score = []
    length = max(len(x), len(y))
    xlen = len(x)
    ylen = len(y)
    for i in range(length):
      diagonal = matrix[xlen-1][ylen-1]
      vertical = matrix[xlen][ylen-1]
      horizontal = matrix[xlen-1][ylen]
      way = max(diagonal, vertical, horizontal)
      if way == diagonal:
        max_score.append(matrix[xlen][ylen])
        max_score.append('diagonal')
        xlen += -1
        ylen += -1
      elif way == vertical:
        max_score.append(matrix[xlen][ylen])
        max_score.append('vertical')
        ylen += -1
      elif way == horizontal:
        max_score.append(matrix[xlen][ylen])
        max_score.append('horizontal')
        xlen += -1
    score = len(max_score)

    k = open("myfile.txt", "w")
    k.write("Sequence1: ")
    k.write(x)
    k.write("\nSequence2: ")
    k.write(y)
    k.write("\nMatch Score: ")
    k.write(str(match_score))
    k.write("\nMismatch Score: ")
    k.write(str(mismatch_score))
    k.write("\nGap Opening Penalty: ")
    k.write(str(gap_opening_penalty))
    k.write("\nGap Extend Penalty: ")
    k.write(str(gap_extend_penalty))
    k.write("\nMaximum Score: ")
    k.write(str(matrix[xlen-1][ylen-1]))
    k.write(" at x: ")
    k.write(str(len(x)))
    k.write(" , y: ")
    k.write(str(len(y)))
    k.write("\n")
    k.write("Alignment Starting From Right Bottom: \n")
    for i in range(score):
     k.write(str(max_score[i]))
     k.write("\n")
    k.write(str(print_matrix(input_file,matrix)))
    k.close()

    print_matrix(input_file,matrix)

global_align(input_file, gap_opening_penalty=-0.5, gap_extend_penalty=-0.1, match_score=2, mismatch_score=-1)
```
